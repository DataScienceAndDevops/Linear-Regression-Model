```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:\Experience-Salary.csv')
data
```

```python
data['exp(in months)'] = data['exp(in months)'].astype(int)
data['salary(in thousands)'] = data['salary(in thousands)'].round(2)
data
```

```python
data.corr()
```

```python
plt.scatter(data['exp(in months)'], data['salary(in thousands)'])
plt.title('Experience vs Salary')
plt.xlabel('Experience (years)')
plt.ylabel('Salary ($)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


X = data[['exp(in months)']]
y = data['salary(in thousands)']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
from sklearn.metrics import r2_score
r_squared = r2_score(y_test, y_pred)
print(f'R-squared: {r_squared}')
```

```python
plt.scatter(X_test, y_test, label='Actual Data')
plt.plot(X_test, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression Model')
plt.xlabel('Experience (years)')
plt.ylabel('Salary ($)')
plt.legend()
plt.show()
```
